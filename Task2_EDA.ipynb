```python
## Age Distribution
Analyzing customer age distribution.
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=10, kde=True)

plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")

plt.show()
```

```python
• Customer ages range from 18 to 65 years.
• Majority of customers belong to the middle-age group.
• Age distribution appears relatively balanced.
```

```python
## Gender Distribution
Analyzing customer gender distribution.
```

```python
df['Gender'].value_counts().plot(kind='bar')

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

```python
• Male and Female customers contribute significantly to sales.
• Customer base is fairly distributed across genders.
```

```python
## Product Category Distribution
Analyzing category-wise records.
```

```python
df['Category'].value_counts().plot(kind='bar')

plt.title("Category Distribution")
plt.xlabel("Category")
plt.ylabel("Count")

plt.show()
```

```python
• Multiple product categories are represented.
• Electronics and other major categories contribute strongly to sales activity.
```

```python
## Top Cities
Cities with highest number of orders.
```

```python
df['City'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Cities")
plt.xlabel("City")
plt.ylabel("Orders")

plt.show()
```

```python
• Certain cities generate more orders than others.
• Customer demand is concentrated in major cities.
```

```python
## Top Products by Revenue
Analyzing products generating highest revenue.
```

```python
top_products = df.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False)

top_products.head(10).plot(kind='bar')

plt.title("Top 10 Products by Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue")

plt.show()
```

```python
• Laptop generated the highest revenue.
• Mobile ranked among the top-performing products.
• Product revenue is concentrated among a few key products.
```

```python
## Revenue by Category
Analyzing category-wise revenue.
```

```python
df.groupby('Category')['Total_Sales'].sum().sort_values().plot(kind='barh')

plt.title("Revenue by Category")
plt.xlabel("Revenue")

plt.show()
```

```python
• Electronics is the highest revenue-generating category.
• Education and Grocery categories also contribute significantly.
• Revenue varies substantially across categories.
```

```python
## Top Cities by Revenue
Analyzing cities generating highest revenue.
```

```python
df.groupby('City')['Total_Sales'].sum().sort_values(ascending=False).head(10).plot(kind='bar')

plt.title("Top 10 Cities by Revenue")
plt.xlabel("City")
plt.ylabel("Revenue")

plt.show()
```

```python
• Patna generated the highest revenue.
• Kolkata and Bengaluru also contributed strongly.
• Revenue distribution differs across cities.
```

```python
## Monthly Sales Analysis
Analyzing sales trends over time.
```

```python
df['Month'] = df['Order_Date'].dt.month_name()
```

```python
monthly_sales = df.groupby('Month')['Total_Sales'].sum()

monthly_sales.plot(marker='o')

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.show()
```

```python
df.groupby('Month')['Total_Sales'].sum().sort_values(ascending=False)
```

```python
• March recorded the highest monthly revenue.
• June and November also showed strong sales performance.
• Sales fluctuate throughout the year.
```

```python
## Correlation Analysis
Analyzing relationships between numerical variables.
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    df[['Age','Quantity','Unit_Price','Total_Sales']].corr(),
    annot=True,
    cmap='Blues'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
• Quantity has a strong positive correlation with Total_Sales (0.65).
• Unit_Price has a strong positive correlation with Total_Sales (0.69).
• Age shows very weak correlation with Total_Sales.
• Sales are mainly influenced by Quantity and Unit_Price.
```
