```python
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[18,25,35,50,65],
    labels=['18-25','26-35','36-50','51-65']
)
```

```python
df[['Age','Age_Group']].head()
```

```python
df.groupby('Age_Group')['Total_Sales'].sum()
```

```python
df.groupby('Age_Group')['Total_Sales'].sum().plot(kind='bar')

plt.title("Revenue by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Revenue")

plt.show()
```

```python
df.groupby('Age_Group')['Order_ID'].count()
```

```python
df.groupby('Age_Group')['Order_ID'].count().plot(kind='bar')

plt.title("Orders by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Orders")

plt.show()
```
